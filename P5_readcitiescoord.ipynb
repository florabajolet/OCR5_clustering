```python
import pandas as pd
import numpy as np
import re
```

```python
regexp = r"<name>(?P<city>[a-z '-]*)<\/name><longitude>(?P<longitude>[0-9-.]*)<\/longitude><latitude>(?P<latitude>[0-9-.]*)<\/latitude>"
```

```python
cities = []
lat = []
lon = []
```

```python
with open(r"E:\Mes Docs\Sciences\Cours\OpenclassroomsDS-ML\Projet5\missing_post_codes.txt", "r", encoding = "utf-8") as f:
    for line in f:
        match = re.match(regexp, line)
        if match is None:
            print(f'no match: {line}')
        else:
            cities.append(match.group('city'))
            lat.append(match.group('latitude'))
            lon.append(match.group('longitude'))
```

```python
coordinates = pd.DataFrame({'customer_city':cities, 'city_lat':lat, 'city_lon':lon})
coordinates
```

```python
coordinates.to_csv('E:/Mes Docs/Sciences/Cours/OpenclassroomsDS-ML/Projet5/coordinates.csv')
```
